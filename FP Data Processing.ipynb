```python
df['symbol'] = df['symbol'].astype(str)
# df['industry'] = industry[df['symbol']]
# df['industry'] = df['symbol'].map(industry.set_index('symbol')['industry'])
df['industry'] = df['symbol'].replace(industries)

df = df.set_index(["industry"])
```

```python
df['expiry_dt'] = pd.to_datetime(df['expiry_dt'])
df['timestamp'] = pd.to_datetime(df['timestamp'])

df['days_to_expiry'] = (df['expiry_dt'] - df['timestamp']).dt.days

df = df[df['days_to_expiry'] <= 365]
```

```python
df['return'] = (df['close'] - df['close'].shift(1)) / df['close']

df = df.dropna()
```
